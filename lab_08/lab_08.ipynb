```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv("penguins.csv")
ds = df[['bill_length_mm', 'flipper_length_mm']]
# plt.scatter(ds['flipper_length_mm'], ds['bill_length_mm'])
```

```python
model = KMeans(n_clusters=2, random_state=False)
model.fit(ds)
# plt.scatter(ds['flipper_length_mm'], ds['bill_length_mm'], c = model.labels_)
# plt.show
```

```python
inertia = []
for k in range(1, 10):
    model = KMeans(n_clusters=k, random_state=False)
    model.fit(ds)
    inertia.append(model.inertia_)
plt.plot(range(1, 10), inertia, marker= 'o')
plt.show()
```

```python
new_ds = df.drop(['rowid', 'species', 'island', 'sex', 'year'], axis = 1)
```

```python
new_model = TSNE(learning_rate=0.001)
new_ds_2d = pd.DataFrame(new_model.fit_transform(new_ds))
```

```python
new_ds_2d_inertia = []
for k in range(1, 10):
    model = KMeans(n_clusters=k, random_state=False)
    model.fit(new_ds_2d)
    new_ds_2d_inertia.append(model.inertia_)
plt.plot(range(1, 10), new_ds_2d_inertia, marker= 'o')
plt.show()
```
